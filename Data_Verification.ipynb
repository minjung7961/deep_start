```python
data_list = df.values[:,:-1] # class 분리
processed_data = [] # 처리된 데이터 저장할 장소

# 더 좋은 방법 있으면 = np.c_ => pandas에서 추가하는 방법. => dataframe을 만들어야될듯 싶음.

# 원시적인 방법
# nparray -> append 달라요 무조건 같은 모양 -> List로 가면 상관없는.
for index, data in enumerate(data_list):
    print(index)
    _temp = data.tolist() #nparray -> list
    _temp.append(norm_amount[0]) # append
    _temp.append(df.values[index][-1]) # class 추가
#     print(len(_temp)) # 길이 확인
    processed_data.append(_temp) # 데이터 저장.

print(processed_data[0][-2]) # 32가 되어야 한다.
# print(len(data_list[0]))
```
